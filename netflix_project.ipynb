```python
# Ensure 'Vote_Count' and 'Vote_Average' are numeric and handle any NaNs
df['Vote_Count'] = pd.to_numeric(df['Vote_Count'], errors='coerce')
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

# Drop rows where 'Vote_Count' or 'Vote_Average' became NaN after coercion
# This is crucial for mean aggregation to work correctly on numeric data
df_cleaned_for_agg = df.dropna(subset=['Vote_Count', 'Vote_Average']).copy()

# Now perform the aggregation on the cleaned DataFrame
df_cleaned_for_agg.groupby('Original_Language').agg({
    'Vote_Count': 'sum',
    'Vote_Average': 'mean'
}).sort_values(by=['Vote_Count', 'Vote_Average'], ascending=False)
```

converted full language names by using language-map

```python
language_map = {
    'en': 'English', 'ja': 'Japanese', 'fr': 'French', 'es': 'Spanish', 'it': 'Italian',
    'ko': 'Korean', 'de': 'German', 'zh': 'Chinese', 'ru': 'Russian', 'sv': 'Swedish',
    'pt': 'Portuguese', 'da': 'Danish', 'hi': 'Hindi', 'pl': 'Polish', 'no': 'Norwegian',
    'id': 'Indonesian', 'th': 'Thai', 'tr': 'Turkish', 'nl': 'Dutch', 'sr': 'Serbian',
    'el': 'Greek', 'ar': 'Arabic', 'nb': 'Norwegian Bokmål', 'ca': 'Catalan', 'lv': 'Latvian',
    'te': 'Telugu', 'ta': 'Tamil', 'ml': 'Malayalam', 'bn': 'Bengali', 'fa': 'Persian',
    'ro': 'Romanian', 'tl': 'Tagalog', 'cs': 'Czech', 'fi': 'Finnish', 'uk': 'Ukrainian',
    'ms': 'Malay', 'hu': 'Hungarian', 'he': 'Hebrew', 'et': 'Estonian', 'la': 'Latin',
    'is': 'Icelandic', 'eu': 'Basque', 'cn': 'Chinese' # 'cn' also maps to Chinese
}

# Normalize 'Original_Language' by stripping whitespace and converting to lowercase
# Then apply the language map using .loc to avoid SettingWithCopyWarning
df.loc[:, 'Original_Language'] = df['Original_Language'].astype(str).str.strip().str.lower().replace(language_map)

# After replacement, some entries might have been mapped to their full names, but if the original
# value was not a key in the map (e.g., if 'zh' was present but not 'ZH'), it remains. We also
# need to convert mapped values to title case or desired display case if necessary.
# Here, we assume the map values are the desired final format.

# Remove any entries that are still unmapped, are URLs, or are None/NaN after initial processing
# We check for 'http' in the string representation and then drop NaNs
df = df[~df['Original_Language'].astype(str).str.contains('http', na=False)].copy() # Explicitly create a copy after filtering
df.dropna(subset=['Original_Language'], inplace=True) # Drop rows where language is None (NaN after string conversion)

# Display unique values and their counts to verify the cleaning
print(df['Original_Language'].unique())
display(df['Original_Language'].value_counts())
```

top 10 movies popularity

```python
top_movies = df.sort_values(by='Popularity', ascending=False).head(10)
plt.figure(figsize=(9, 5))
plt.barh(top_movies['Title'], top_movies['Popularity'])
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies by Highest Popularity')
plt.tight_layout()
plt.show()
```

top languages Analysis-pie chart

```python
import matplotlib.pyplot as plt
import pandas as pd # Ensure pandas is imported

# Calculate the top 5 original languages
language_counts = df['Original_Language'].value_counts().head(5)

plt.figure(figsize=(7,4))

plt.pie(
    language_counts, # Use the correct variable for language counts
    labels=language_counts.index,
    autopct='%1.1f%%'
)

plt.title('Top 5 Original Languages') # Corrected title for clarity
plt.legend(loc='best', bbox_to_anchor=(1,1))
plt.show()
```

pie -chart (top 5 genre)

```python
# Top 5 Genres
genre_counts = df['Genre'].value_counts().head(5)

plt.figure(figsize=(7,4))

plt.pie(
    genre_counts,
    labels=genre_counts.index,
    autopct='%1.1f%%'
)

plt.title('Top 5 Genres')
plt.legend(loc='best', bbox_to_anchor=(1,1))
plt.show()
```

movies release_Date per year

```python
df['Year'] = pd.to_datetime(df['Release_Date']).dt.year

movies_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(
    movies_per_year.index,
    movies_per_year.values,
    color='red',
    marker='o'
)

plt.title('Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)

plt.show()
```

Rating Distribution-->histrogram

```python
import pandas as pd


df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

# Drop any NaN values from the 'Vote_Average' Series to ensure the data for the histogram
# contains only valid numerical entries. This will prevent any `None` (if they somehow exist)
# or `NaN` values from interfering with matplotlib's internal type checking.
data_to_plot = df['Vote_Average'].dropna()

plt.figure(figsize=(8,5))

plt.hist(data_to_plot,
         bins=10,
         color='blue')

plt.title('Rating Distribution')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')

plt.show()
```
